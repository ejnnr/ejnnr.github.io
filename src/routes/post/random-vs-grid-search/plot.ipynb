```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
xx, yy = np.meshgrid(np.linspace(0.05, 0.95, 9), np.linspace(0.05, 0.95, 9))
xx = xx.flatten()
yy = yy.flatten()
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(xx, yy, s=4)
plt.title("Grid")
plt.axis("square")
plt.xticks([])
plt.yticks([])
plt.xlim(0, 1)
plt.ylim(0, 1)

plt.subplot(1, 2, 2)
x_rand = np.random.rand(81)
y_rand = np.random.rand(81)
plt.scatter(x_rand, y_rand, s=4)
plt.title("Random")
plt.axis("square")
plt.xticks([])
plt.yticks([])
plt.xlim(0, 1)
plt.ylim(0, 1)
```

```python
xx, yy = np.meshgrid(np.linspace(0.05, 0.95, 9), np.linspace(0.005, 0.095, 9))
xx = xx.flatten()
yy = yy.flatten()
plt.figure(figsize=(10, 4))
plt.subplot(2, 1, 1)
plt.scatter(xx, yy, s=4)
plt.title("Grid")
plt.gca().set_aspect("equal")
plt.xlim(0, 1)
plt.ylim(0, 0.1)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 1, 2)
x_rand = np.random.rand(81)
y_rand = 0.1 * np.random.rand(81)
plt.scatter(x_rand, y_rand, s=4)
plt.title("Random")
plt.gca().set_aspect("equal")
plt.xlim(0, 1)
plt.ylim(0, 0.1)
plt.xticks([])
plt.yticks([])
```
